# Count the digits

```python
# The long method with more time complexity
def count_dig():
    n=int(input("Enter a number: "))
    count=0
    while(n>0):
        count=count+1
        n=n//10
    print("Number of digits:",count)
# Calling the function
count_dig()
```

```python
# Similarly we can directly use the built-in function to find the number of digits
def count_dig_builtin():
    n=int(input("Enter a number: "))
    print("Number of digits:",len(str(n)))
# Calling the function
count_dig_builtin()
```

```python
# The short method with less time complexity
def count_dig_short():
    n=int(input("Enter a number: "))
    import math
    if n>0:
        digits=math.floor(math.log10(n))+1
    
    print("Number of digits:",digits)
# Calling the function
count_dig_short()
```

# Reversing the number

```python
# Reversing the number
n=int(input("Enter a number to reverse: "))
rev=0
last_digit=0
while(n>0):
    last_digit=n%10
    rev=(rev*10)+last_digit
    n=n//10
print("Reversed Number:",rev)
```

# Check pallindrome

```python
# Reversing the number first and then checking for palindrome


n=int(input("Enter a number to reverse: "))
rev=0
cpy=n
last_digit=0
while(n>0):
    last_digit=n%10
    rev=(rev*10)+last_digit
    n=n//10

if (rev==cpy):
    print("Palindrome Number") 
else:
    print("Not a Palindrome Number")    
```

# Armstrong number (for only 3 digits)

```python
n1=int(input("Enter the number: "))
sum=0
last=0,0
copy=n1
while(n1>0):
    last=n1%10
    sum=sum+(last**3)                    # sum of cubes of each digit = Armstrong condition
    n1=n1//10                            # 371 = 3**3 + 7**3 + 1**3 = 371

if (sum==copy):
    print("Armstrong Number")
else:        
    print("Not an Armstrong Number")
```

```python
# WORKS ON ALL ARMSTRONG NUMBERS

n1 = int(input("Enter the number: "))
digits = len(str(n1))             # Counting number of digits
total = 0                         # If 4 digits then sum of 4th power of each digit and so on
copy = n1

while n1 > 0:
    last = n1 % 10
    total += last ** digits
    n1 //= 10

if total == copy:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

# GCD
Greatest Common Factor (GCF), is the largest positive integer that divides two or more integers without leaving a remainder

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

gcd = 1
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        gcd = i

print("GCD is:", gcd)
```

# Prime number

```python
n=int(input("Enter a number: "))
count=0
for i in range (1,n+1):          # n+1 coz we want to include n in the range i.e 1 and the number itself
    if (n%i==0):
        count=count+1
if (count==2):                   #1 and the number itself are the only factors of a prime number
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
# Reversing an array using loop
n=int(input("Enter number of elements in the array: "))
arr=[]
for i in range(0,n):
    ele=int(input("Enter the element : "))
    arr.append(ele)
    rev_Arr=[]
for j in range(n-1,-1,-1):
    rev_Arr.append(arr[j])
print("Reversed Array is : ",rev_Arr)
```

```python
# checking if a string is palindrome or not
s=input("Enter a string: ") 
rev_s=""
for i in range(len(s)-1,-1,-1):
    rev_s=rev_s+s[i]    
if (s==rev_s):
    print("Palindrome String")  
else:   
    print("Not a Palindrome String")
```
